```python
q=data.iloc[:,3:].values
q
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=3,min_samples=4,metric='euclidean')
```

```python
m=db.fit(q)
m
```

```python
m.labels_
```

```python
m=db.fit_predict(q)
plt.scatter(q[m==0,0],q[m==0,1],color='red',label='cluster1')
plt.scatter(q[m==1,0],q[m==1,1],color='green',label='cluster2')
plt.scatter(q[m==2,0],q[m==2,1],color='orange',label='cluster3')
plt.scatter(q[m==3,0],q[m==3,1],color='blue',label='cluster4')
plt.scatter(q[m==4,0],q[m==4,1],color='purple',label='cluster5')
plt.scatter(q[m==5,0],q[m==5,1],color='yellow',label='cluster6')
plt.scatter(q[m==6,0],q[m==6,1],color='black',label='cluster7')
plt.scatter(q[m==7,0],q[m==7,1],color='pink',label='cluster8')
plt.scatter(q[m==8,0],q[m==8,1],color='red',label='cluster9')
```

```python
data['cluster']=pd.DataFrame(m)
```
